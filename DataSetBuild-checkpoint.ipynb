```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read your dataset

# Select relevant features for the heatmap
selected_features = [
    'age', 'g', 'gs', 'mp_per_g', 'fg_per_g', 'fga_per_g', 'fg_pct', 'fg3_per_g', 'fg3a_per_g',
    'fg3_pct', 'fg2_per_g', 'fg2a_per_g', 'fg2_pct', 'efg_pct', 'ft_per_g', 'fta_per_g', 'ft_pct',
    'orb_per_g', 'drb_per_g', 'trb_per_g', 'ast_per_g', 'stl_per_g', 'blk_per_g', 'tov_per_g',
    'pf_per_g', 'pts_per_g', 'per', 'ts_pct', 'fg3a_per_fga_pct', 'fta_per_fga_pct', 'orb_pct',
    'drb_pct', 'trb_pct', 'ast_pct', 'stl_pct', 'blk_pct', 'tov_pct', 'usg_pct', 'ows', 'dws',
    'ws', 'ws_per_48', 'obpm', 'dbpm', 'bpm', 'vorp', 'award_share', 'mov', 'mov_adj'
]

# Create a subset of the dataset with selected features
subset = player_dataset[selected_features]

# Calculate the correlation matrix
correlation_matrix = subset.corr()

# Set the threshold for correlation
correlation_threshold = 0.7  # You can adjust this threshold as needed

# Find highly correlated pairs of features
highly_correlated_pairs = (correlation_matrix.abs() > correlation_threshold) & (correlation_matrix != 1)

# Create a subset of the correlation matrix for highly correlated features
subset_corr_matrix = correlation_matrix[highly_correlated_pairs]

# Create a heatmap using Seaborn for the subset
plt.figure(figsize=(20, 16))
sns.heatmap(subset_corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Highly Correlated Features')
plt.show()
```

```python
filtered_players = player_dataset[
    ((player_dataset['g'] > 48) & (player_dataset['pts_per_g'] > 13.5) & (player_dataset['mp_per_g'] > 30) &
     (player_dataset['team_id'] != 'TOT') & (player_dataset['fg_pct'] > 0.37) & (player_dataset['fga_per_g'] > 10) &
     (player_dataset['per'] > 18)) | (player_dataset['award_share'] > 0)
].reset_index(drop=True)

print(filtered_players)
```

```python
merged_data = pd.merge(filtered_players, df[['full_name', 'id']], how='left', left_on='player', right_on='full_name')

# Drop the redundant 'full_name' column
merged_data = merged_data.drop(columns=['full_name'])

# Display the result
print(merged_data)
```

```python
def find_player_id(player_name, data1):
    """
    Find the player ID based on the player name in data1.
    
    Parameters:
    - player_name (str): The name of the player to search for.
    - data1 (pd.DataFrame): The DataFrame containing player information.

    Returns:
    - int or None: The player ID if found, otherwise None.
    """
    player_id = data1.loc[data1['first_name'] == player_name, 'id'].values
    return player_id[0] if len(player_id) > 0 else None
# Example usage
player_name_to_search = 'World'
player_id = find_player_id(player_name_to_search, df)

if player_id is not None:
    print(f"The ID of {player_name_to_search} is {player_id}.")
else:
    print(f"Player {player_name_to_search} not found in data1.")
```
